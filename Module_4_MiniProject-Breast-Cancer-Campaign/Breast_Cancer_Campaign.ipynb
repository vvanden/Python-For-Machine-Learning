# Step 4

Clustering
Use k-means and Mean-shift to fit your data.
Get optimal k for k-means using the elbow curve method.
Try different bandwidth for the mean-shift clustering.

```python
from sklearn.cluster import KMeans

distorions = []
spread = range(1,26)

for k in spread:
    model = KMeans(n_clusters=k, tol=0.01, random_state=0)
    model.fit(df_ins_1)
    distorions.append(model.inertia_)

print(distorions)
```

```python
import matplotlib.pyplot as plt
plt.plot(spread, distorions, marker = ".", c="red")
plt.title("Elbow Finder - INS")
plt.xlabel("Np. of Clusters")
plt.ylabel("SSE (Model Inertia)")
plt.show()
```

```python
distorions = []
spread = range(1,26)

for k in spread:
    model = KMeans(n_clusters=k, tol=0.01, random_state=0)
    model.fit(df_wis_1)
    distorions.append(model.inertia_)

print(distorions)
```

```python
import matplotlib.pyplot as plt
plt.plot(spread, distorions, marker = ".", c="blue")
plt.title("Elbow Finder - WIS")
plt.xlabel("Np. of Clusters")
plt.ylabel("SSE (Model Inertia)")
plt.show()
```
